```python
import pandas as pd
import numpy as np


data = pd.read_csv('2017_jun_final.csv')
data.head()
```

```python
data.shape
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data.drop(["Специализация", "Университет", "Предметная.область", "Валюта", "cls"], axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data_new = data.groupby(['Должность']).mean(numeric_only=True)
data_new.head()
```

```python
data_new.agg({'Зарплата.в.месяц' : ['min', 'max']})
```

```python
def fill_avg_salary():
    avg = np.mean(data_new['Зарплата.в.месяц'])
    return(f'%.2f' % avg)
    
avg = data_new.apply(lambda x: fill_avg_salary(), axis=1)
data_new['avg'] = avg
data_new.head()
```

```python
data_new.describe()
```

```python
data_new.to_csv('2023_final.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
sns.set(font_scale=1)

data_new.sort_values(by='exp', inplace=True)

plt.plot(
    data_new['exp'],
    data_new['salary'],
)
plt.title("Графік залежності заробітної плати від досвіду роботи", fontsize=12, color="b")
plt.xlabel("Досвід роботи", fontsize=10, color="g")
plt.ylabel("Заробітна плата", fontsize=10, color="g")
```

```python
sns.set_style("whitegrid")
data_new.sort_values(by='Возраст', inplace=True)

plt.scatter(
    data_new['Возраст'],
    data_new['salary'],
)
plt.title("Залежность заробітної плати від віку", fontsize=12, color="b")
plt.xlabel("Досвід роботи", fontsize=10, color="g")
plt.ylabel("Заробітна плата", fontsize=10, color="g")
```

```python
sns.set_style("darkgrid")
sns.set(font_scale=0.2)
sns.stripplot(x='Должность', y='Зарплата.в.месяц', data=data)
plt.title("Залежність заробітної плати від посади", fontsize=15, color="b")
plt.xlabel("Должность", fontsize=10, color="g")
plt.ylabel("Зарплата.в.месяц", fontsize=10, color="g")
```

```python
sns.set_style("whitegrid")
sns.set(font_scale=0.5)
sns.relplot(x='Общий.опыт.работы', y='Зарплата.в.месяц', kind='scatter', hue='Должность', data=data)
plt.title("Залежність заробітної плати від досвіду роботи по посадах", fontsize=10)
```
